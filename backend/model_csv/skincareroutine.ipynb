```python
import pandas as pd
import numpy as np
import lightgbm as lgb
import sklearn

print("✅ Packages installed successfully!")
```

```python
# =========================
# Load dataset
# =========================
import os

file_path = r"D:\sem 7 project\backend\model_csv\skincare_dataset_500.csv"
df = pd.read_csv(file_path)

df.head()  # Preview first 5 rows
```

```python
# =========================
# Define features (X) and labels (Y)
# =========================
X = df[["Age", "Gender", "Skin_Type", "Smoking", "Alcohol", "Diabetes",
        "Diabetes_Type", "Location", "Common_Concern", "Pollution_Level"]]

Y = df[["morning_cleanser","morning_toner","morning_serum","morning_moisturizer",
        "morning_sunscreen","morning_exfoliator","morning_mask",
        "night_cleanser","night_toner","night_serum","night_moisturizer",
        "night_exfoliator","night_mask"]]

X.head(), Y.head()
```

```python
# =========================
# Preprocess categorical features
# =========================
categorical_cols = ["Skin_Type", "Gender", "Smoking", "Alcohol", "Diabetes",
                    "Diabetes_Type", "Location", "Common_Concern", "Pollution_Level"]

# Convert categorical columns to string
for col in categorical_cols:
    X[col] = X[col].astype(str)

# One-hot encode categorical features
X_encoded = pd.get_dummies(X)

X_encoded.head()
```

```python
# =========================
# Split data into training and testing sets
# =========================
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X_encoded, Y, test_size=0.2, random_state=42
)

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
# =========================
# Train LightGBM MultiOutput model
# =========================
from sklearn.multioutput import MultiOutputClassifier
import lightgbm as lgb

lgb_model = MultiOutputClassifier(lgb.LGBMClassifier())
lgb_model.fit(X_train, Y_train)

print("✅ Model training completed!")
```

```python
# =========================
# Save trained model
# =========================
import pickle

output_dir = r"D:\sem 7 project\backend\modelsoutput"
os.makedirs(output_dir, exist_ok=True)

save_path = os.path.join(output_dir, "skincare_lgb_model.pkl")
with open(save_path, "wb") as f:
    pickle.dump((lgb_model, X_encoded.columns.tolist()), f)

print(f"✅ Model saved at {save_path}")
```

```python
# =========================
# Quick test of the trained model
# =========================
sample = X_encoded.iloc[0:1]
pred = lgb_model.predict(sample)
pred
```
